```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
path_to_cleaned_data = "cleaned.csv"
df = pd.read_csv(path_to_cleaned_data)
```

```python
df["title_len"] = df['title'].str.len()
df['contains_subheading'] = df['title'].str.contains(":",regex=False)
df["title_num_words"] = df['title'].str.split().str.len()
df
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = df[["title_len"]].values
y = df[["review_score"]].values
model = LinearRegression().fit(x, y)
model.score(x, y)
```

```python
y_pred = model.predict(x)
plt.scatter(x, y, color ='b')
plt.plot(x, y_pred, color ='k')
plt.show()
```
